Drop the unnecessary column

```python
df.drop("LoanID",axis = 1, inplace=True)
```

Label encoding for 'Education' column

```python
df['Education'].unique()
```

```python
df['Education'] = df['Education'].map({"High School": 0, "Bachelor's": 1, "Master's": 2, "PhD": 3})
```

```python
df['Education'].unique()
```

One hot encoding for nominal columns

```python
df_ohe = pd.get_dummies(df, columns=["EmploymentType", "MaritalStatus", "LoanPurpose", "HasMortgage", "HasDependents", "HasCoSigner"])
```

```python
df_ohe.shape
```

Using SMOTE for balancing the classes

```python
x = df_ohe.drop(columns=['Default'])
y = df_ohe['Default']
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=10)

x_resample, y_resample = smote.fit_resample(x, y)

resample_df = pd.DataFrame(x_resample, columns=x.columns)
resample_df["Default"] = y_resample
```

```python
resample_df.to_csv('Loan_default_Preprocessing.csv', index = False)
```
